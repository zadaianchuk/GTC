4. What does GLM stand for?
> Generalized Linear Model
5. What is the general idea behind GLM?
> We have different predictors $X_j$, and assume that our signal is weighted sum of these predictors and some  Gaussian noise $\varepsilon$
> Our aim is to find weights $\beta$ that minimize least square error $S=\varepsilon^{T}\varepsilon=\sum_i (Y_i-\hat{Y_i})^2 $
6. What is the basic equation?
> $Y=X\beta +\varepsilon = \hat{Y} +\varepsilon$
7. What are advantages of using a GLM analysis?
> - any shape of expected time course (predictors are continuous)
> - many predictors (you can add noise predictors to cancel them out)

8. Which other tests are special cases of a GLM?
> one-sample t-tests, 2-sample t-tests, paired t-test, ANOVA, ANCOVA
