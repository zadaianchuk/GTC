b) PCA

```python
X = np.concatenate((X0, X1), axis=1);
mean = np.mean(X, axis=1);
X = (X.transpose() - mean).transpose();
[U, S, V] = np.linalg.svd(np.divide(np.dot(X, X.transpose()), X.shape[1]));
s_sum = np.cumsum(S);
```

Cumulative sum of singular values

```python
plt.plot(s_sum);
```

```python
threshold = 0.9*s_sum[-1];
m = (s_sum <= threshold).sum();
Y = np.dot(U[:,0:m].transpose(), X);
X_rec = np.dot(U[:,0:m], Y);
```

```python
m
```

Reconstruction error

```python
np.mean(np.square(X_rec - X))/np.mean(np.square(X))
```

c) Fraction of energy for two first PC

```python
s_sum[1]/s_sum[-1]
```

error of reconstraction

```python
Z = np.dot(U[:,0:2].transpose(), X);
X_rec2 = np.dot(U[:,0:2], Z);
np.mean(np.square(X_rec2 - X))/np.mean(np.square(X))
```

c) Download test data

```python
test = scipy.io.loadmat('mnist_test.mat');
test = test['test'];
X0_test = test[0,9];
X1_test = test[0,0];
```

```python
X_test = np.concatenate((X0_test, X1_test), axis=1);
mean_test = np.mean(X_test, axis=1);
X_test = (X_test.transpose() - mean_test).transpose();
Z_test = np.dot(U[:,0:2].transpose(), X_test);
Z0_test = Z_test[:, 0:X0_test.shape[1]]
Z1_test = Z_test[:, X0_test.shape[1]:Z_test.shape[1]]
```

d)

```python
from numpy.linalg import inv
```

fit linear regression

```python
C0 = np.zeros(Z0.shape[1]);
C1 = np.ones(Z1.shape[1]);
C = np.append(C0, C1);
w = 10000*np.dot(np.dot(C, Z.transpose()), inv(np.dot(Z, Z.transpose())));
threshold = -1100;
```

```python
c_pred = 0.5*(np.sign(np.dot(w, Z)+threshold)+1)
Q = ((c_pred == C).sum())/Z.shape[1]
```

```python
C0_test = np.zeros(Z0_test.shape[1]);
C1_test = np.ones(Z1_test.shape[1]);
C_test = np.append(C0_test, C1_test);
c_pred_test = 0.5*(np.sign(np.dot(w, Z_test)+threshold)+1)
Q_test = ((c_pred_test == C_test).sum())/Z_test.shape[1]
```

e)

```python
F = np.zeros([5,Z.shape[1]])
for i in range(0,Z.shape[1]):
    F[0,i] = Z[0,i];
    F[1,i] = Z[1,i];
    F[2,i] = Z[0,i]*Z[0,i];
    F[3,i] = Z[1,i]*Z[1,i];
    F[4,i] = Z[0,i]*Z[1,i];
```

```python
w_F = 10000*np.dot(np.dot(C, F.transpose()), inv(np.dot(F, F.transpose())));
threshold_F=-4480;
```

```python
c_pred_F = 0.5*(np.sign(np.dot(w_F, F)+threshold_F)+1)
Q = ((c_pred_F == C).sum())/F.shape[1]
```

```python
F_test = np.zeros([5,Z_test.shape[1]])
for i in range(0,Z_test.shape[1]):
    F_test[0,i] = Z_test[0,i];
    F_test[1,i] = Z_test[1,i];
    F_test[2,i] = Z_test[0,i]*Z_test[0,i];
    F_test[3,i] = Z_test[1,i]*Z_test[1,i];
    F_test[4,i] = Z_test[0,i]*Z_test[1,i];
```

```python
c_pred_F_test = 0.5*(np.sign(np.dot(w_F, F_test)+threshold_F)+1)
Q_test = ((c_pred_F_test == C_test).sum())/F_test.shape[1]
```
